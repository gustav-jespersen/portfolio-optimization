# Geometric Brownian Motion

Geometric Brownian Motions stems from biology and is used to model a random walk with constant mean and variance. Using it for stock prices, it takes the following form.

$$
\frac{S_{t+1} - S_t}{S_t} = \mu \Delta t + \sigma \epsilon \sqrt{\Delta t}
$$

* $S$: Asset price.
* $\mu$: Drift of asset.
* $\sigma$: Volatility of asset.
* $\epsilon$: Normal random variable.

In essence the future stock price is simply current price first multiplied by its drift and second by its volatility, which is the stochastic part, where the element of $\epsilon$ is the stochasticity.

# Simulating correlated returns

**Protocol**

1. Simulate uncorrelated returns.<br>
2. Select random assets from simulated returns.<br>
3. Simulate new uncorrelated returns on-top of random selected assets.<br>
     3.1) Take the simulated returns and add the new simulated returns to them, in that manner creating correlation with original returns.<br>
4. Append the new uncorrelated returns to the original return ~ thus creating correlation in the returns.

**Parameters**

* ```mu```: Mean return in simulation.
* ```sigma```: Standard deviation in simulation. 
* ```nObs```: Number of observations per asset.
* ```size```: Number of assets.
* ```sigmaF```: Fraction of standard deviation for correlated assets.

````python
mu = 0 # mean return
sigma = 0.01 # standard deviation
nObs = 520 # number of observations per asset
size = 10 # number of assets
sigmaF = 0.25 # the fraction of variance between original assets and correlated assets

# 1) original uncorrelated assets
orgSize = int(0.5 * size) # the number of original assets ~ half size of total number of assets
x = np.random.normal(mu, sigma, size=(nObs, orgSize))

# 2) selecting random assets for correlation
corrAssets = [random.randint(0, orgSize-1) for i in range(orgSize)]

# 3) creating new returns which are correlated with corrAssets
y = x[:, corrAssets] + np.random.normal(0, sigma*sigmaF, size = (nObs, len(corrAssets)))

# 4) appending correlated returns to original returns
x = np.append(x, y, axis=1)
````

````python
def correlated_returns(mu, sigma, nObs, size, sigmaF):
    
    # original uncorrelated assets
    orgSize = int(0.5 * size) # the number of original assets ~ half size of total number of assets
    x = np.random.normal(mu, sigma, size=(nObs, orgSize))

    # selecting random assets for correlation
    corrAssets = [random.randint(0, orgSize-1) for i in range(orgSize)]

    # creating new returns which are correlated with corrAssets
    y = x[:, corrAssets] + np.random.normal(0, sigma*sigmaF, size = (nObs, len(corrAssets)))

    # appending correlated returns to original returns
    x = np.append(x, y, axis=1)
    
    return x
````

### Plotting correlated returns (no shocks)

````python
ret = x

fig = plt.figure(figsize=(15,10))

for asset in range(0,len(ret[0])):
    
    cum_ret = (pd.Series(ret[:,asset])+1).cumprod()
    periods = np.arange(0, len(ret))

    plt.plot(periods, cum_ret, label=("Asset " + str(asset)))


plt.title("Correlated assets cumulative return", size=20)
plt.xlabel("Period", size=16)
plt.ylabel("Cumulative return", size=16)

plt.grid(linestyle="--")
plt.legend()
plt.show()
````

### Adding random common shocks ~ RCS

Adding a random common shock simply means to select two random, but correlated, returns and add a shock in each direction.

**Protocol**
1) Random selection of index for adding shocks.
2) Adding shocks to correlated assets.

**Parameters**
* ```sLength```: The number of observations into the full period, at which the shock may occur.
    * Thus, 0 < ```sLength``` < ```nObs```


````python
sLength = 260

point = np.random.randint(sLength, nObs-1, size=2)
x[np.ix_(point, [corrAssets[0], orgSize])] = np.array([[-.5, -.5], [2,2]])

print("Shock at:", min(point), "for assets:", corrAssets[0], "& 5")
print("Reverse shock at:", max(point), "for assets:", corrAssets[0], "& 5")
````

````python
ret = x

fig = plt.figure(figsize=(15,10))

for asset in range(0,len(ret[0])):
    
    cum_ret = (pd.Series(ret[:,asset])+1).cumprod()
    periods = np.arange(0, len(ret))

    plt.plot(periods, cum_ret, label=("Asset " + str(asset)))


plt.title("Correlated assets cumulative return (w. RCS)", size=20)
plt.xlabel("Period", size=16)
plt.ylabel("Cumulative return", size=16)

plt.grid(linestyle="--")
plt.legend()
plt.show()
````

### Adding random specific shock ~ RSS

````python
point = np.random.randint(sLength, nObs-1,size=2)
x[point, corrAssets[-1]] = np.array([-.5,2])
````

````python
ret = x

fig = plt.figure(figsize=(15,10))

for asset in range(0,len(ret[0])):
    
    cum_ret = (pd.Series(ret[:,asset])+1).cumprod()
    periods = np.arange(0, len(ret))

    plt.plot(periods, cum_ret, label=("Asset " + str(asset)))


plt.title("Asset cumulative return (w. RCS and RSS)", size=20)
plt.xlabel("Period", size=16)
plt.ylabel("Cumulative return", size=16)

plt.grid(linestyle="--")
plt.legend()
plt.show()
````
