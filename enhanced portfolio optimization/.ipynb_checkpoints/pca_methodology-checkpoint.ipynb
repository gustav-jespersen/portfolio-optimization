# Principal component analysis

This notebook contains a simple exploration of and introduction to principal component analysis. This is done for the purpose of understanding the Enhanced Portfolio Optimization (EPO) model where it is used to identify the 'problem portoflios'.

First I will introduce the methodology for a manual computation, but in the example below, I will simply use the sklearn module.

## Steps of computing principal components

<b>Step 1: Covariance matrix</b><br>

The first step is to calculate the covariance matrix of the stock returns.

<b>Step 2: Eigenvectors and -values</b><br>

The second step is to calculate the eigenvectors and -values of the covariance matrix.

$$
A \nu = \lambda \nu
$$

Where $A$ is the matrix of returns (or in other cases just any matrix of variables), $\nu$ the eigenvectors, and $\lambda$ the eigenvalues.

<b>Step 3: Sort by eigenvalues</b><br>

The next step is to order the eigenvalues in descending order. Here, it is important to be able to access the corresponding eigenvector for each eigenvalue after the sorting.

<b>Step 4: Principal components</b><br>

The fourth step is to calculate how much each principal components explains of the variance in the data. This is done by dividing each sorted eigenvalue by the sum of the eigenvalues.

$$
\nu_i = \frac{\lambda_i}{\sum^{N}_{i=1}{(\lambda)}}
$$

Where $N$ is the number of variables in matrix $A$. Since the eigenvalues are sorted descendingly, PC1 will contain the higest percentage of explanation and so on.

<b>Step 5: Feature vector</b><br>

The feature vector is where dimensionality reduction is performed. Here, one determine which eigenvalues to keep depending on their variance explenation percentage from Step 4. From there the feature vector is constructed with the eigenvectors as columns corresponding to the eigenvalues choosen.

$$
B = 
\begin{bmatrix}
\nu_{1.1} & \nu_{2.1} & \dots & \nu_{N.1}\\
\nu_{1.2} & \nu_{2.2} & \dots & \nu_{N.2}\\
\vdots    & \vdots    & \vdots& \vdots\\
\nu_{1.T} & \nu_{2.T} & \dots & \nu_{N.T}\\
\end{bmatrix}
$$

Where $\nu_{1}$ is the first eignevector and $N$ is the last eigenvector corresponding to the last selected eigenvalue from Step 4. Whether or not to keep eigenvalues might be determined after examining the eigenvalues. There might be a different cut-off depending on the structure of the decrease. A linear decrease might have a different cut-off compared to an exponential decrease.

# Example using stock returns

<u>Joblist</u>
1. Obtain stock daily returns.
2. Calculate PCs using sklearn.
3. Determine weight of each stock in each PC.
4. Calculate returns of equal weighted and PC1 portfolios.
5. Compare the results for each portfolio.

### (1) Return data

I use a function located in the codelib/structure_functions, which simply imports adjusted daily returns per ticker within a time period. Also it drops rows for any cell with NaN value ~ the function prints how many rows where dropped.

````python
# configuration of data pull
start = "2020-01-01"
end = "2021-12-31"
tickers = ['AAPL','MSFT','AMZN','TSLA','IBM','SPY','DIS','COST','TGT','DAL']

# codelib/structure_functions.daily_returns()
returns_df = daily_return(start, end, tickers)
returns = returns_df.to_numpy()
````

### (2) Principal components

In this step I use the sklearn module to compute the principal components. The function ```PCA()``` also alows for extracting the  *explained variance ratio*, which as the name suggests, is the ratio of explained variance per portfolio.

````python
# sklearn for PC
n = len(tickers)
sklearn_pca = PCA(n_components=n)
pc = sklearn_pca.fit_transform(returns)

# plotting variance explenation ratio
plt.figure(figsize=(10,5))
plt.bar(range(n), sklearn_pca.explained_variance_ratio_, label="Variance ratio per PC")
plt.title('Variance explained by PC', size=14)
plt.grid(linestyle="--")
plt.legend()
plt.show()
````

<b>Comments</b><br>
It is apparent that PC1 explains roughly 50% of the variance in the original data and from there an exponential decline in explenation.

### (3) Obtain weights of stocks for each PC

In this step I simply obtain the stock weights for each PC portfolio.

````python
# components
pcs = sklearn_pca.components_

# create matrix of weights per PC
weights = np.empty([len(pcs),n])

idx = 0
for pc in pcs:
    pc_weight = np.array(pc / sum(pc).T)
    weights[idx] = pc_weight
    idx += 1
````

### (4) Calculate returns for equal weighted and PC1

For calculating the equal weighted returns, I use a function from codelib/structure_functions named ```equal_weighted_portfolio()```, which simply takes in an array of returns (columns is stock return series), and returns an new array of equal weighted portfolio.

````python
# returns of PC1
pc_num = 0
pc_returns = np.matmul(returns, weights[pc_num])
cum_pc_returns = (pc_returns + 1).cumprod()

# equal weighted portfolio
eq_returns = equal_weighted_portfolio(returns)
cum_eq_returns = (eq_returns + 1).cumprod()

# difference
diff = pc_returns - eq_returns
````

### (5) Compare results of portfolios

````python
fig, axs = plt.subplots(2, 1, figsize=(15,10), gridspec_kw={'height_ratios': [4, 2]})

# setting sizes
ax = axs[0].twinx()

# x-values
x_val = returns_df.index

# upper suplot
axs[0].plot(x_val, cum_pc_returns, label="Cumulative PC1 daily returns")
axs[0].plot(x_val, cum_eq_returns, label="Cumulative equal weighted daily returns")
ax.plot(x_val, pc_returns, color="red", label="PC1 daily returns")
ax.plot(x_val, eq_returns, color="black", label="Equal weighted daily returns")

# lower suplot
axs[1].bar(x_val, diff, label="Difference in daily returns")

# axis layout
axs[0].grid(linestyle="--")
axs[1].grid(linestyle="--")

ax.set_ylim(-0.3, 0.3)
ax.set_xlabel("Date")
axs[0].set_ylabel("Cumulative daily returns", size=12)
ax.set_ylabel("Daily returns", size=12)
axs[1].set_ylabel("Difference in daily returns", size=12)

axs[0].legend()
axs[1].legend()
ax.legend()

axs[0].set_title("Cumulative and Daily Returns", size=14)
axs[1].set_title("Difference in daily returns", size=14)

plt.show()
````

<b>Comments</b><br>

From subplot one it is clear that the daily returns move highly similar. This is not to be mistaken with the cumulative returns, where there is a significant difference. This is due to the sum of the difference, which is positive in this case (values of 0.27), which indicates PC1 has summed more positive returns compared to the equal weighted portfolio. <br>

From subplot two is is apparant and more clearly stated when the portfolios diverge from each other.
